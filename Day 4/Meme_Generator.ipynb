# Data Collection

Spraping data from "imgfli.com"

** Scraping from Imgflip Meme Gallery **

```python
import requests
from bs4 import BeautifulSoup

# Data Scinece Memes Url
url = "https://imgflip.com/memetemplates"

# Sending get request
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Finding meme templates
memes = []
for img in soup.find_all('img', class_="shadow"):
    memes.append({
        "alt" : img.get('alt'),
        "src" : img.get('src')
        })

# Scraped Memes:
for meme in memes:
    print(meme)
```

**Dowloading Meme Images**

```python
import os

# Folder to save memes
if not os.path.exists("memes"):
    os.makedirs("memes")

# Downloading memes
for i, meme in enumerate(memes):
    try:
      response = requests.get(meme["src"])
      if response.status_code == 200:
        with open(f"memes/meme_{i}.jpg", "wb") as f:
          f.write(response.content)
        print(f"Downloaded meme_{i}: {meme['alt']}")
      else:
        print(f"Failed to download meme_{i}: Status code {response.status_code}")
    except Exception as e:
      print(f"Error downloading meme_{i}: {e}")
```

# DATA PREPROCESSING

Preprocessing the data to make it ready for AI model.

- **Resize Images**: Resize images to maintain consistency
- **Clean Text**: Clean the 'alt' text

```python
from PIL import Image

# Resize images
for i in range(len(df)):
  try:
    img_path = f"memes/meme_{i}.jpg"
    img = Image.open(img_path)
    img = img.resize((256, 256))
    img.save(img_path)
    print(f"Resized image {i}")
  except Exception as e:
    print(f"Error resizing image {i}: {e}")
```

```python
import re

# Cleaning alt text
def clean_text(text):
  if isinstance(text, str):
    text = re.sub(r"[^\w\s]", "", text) # remove special characters
    return text.strip()
  return ""

df["alt"] = df["alt"].apply(clean_text)

# Saving the cleaned DataFrame
df.to_csv("memes_cleaned.csv", index=False)
```

# Traing AI Model

Since the dataset is ready, we can start working on our ai model

**Text Generation**
- Use the captions to train an NLP model for generating funny captions.

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TFTrainingArguments

# Loading pre-trained GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Tokenizing the captions
tokenized_text = [tokenizer(text, return_tensors="pt") for text in df["alt"]]

# Fine tuning the model
training_args = TFTrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=2,
    save_steps=10_000,
    save_total_limit=2,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_text,
)

trainer.train()
```

**Image Generation**
- Usining meme images to train a GAN Model for generating new meme templates

```python
# Loading images for training a GAN
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset

class MemeDataset(Dataset):
  def __init__(self, image_folder, transform=None):
    self.image_folder = image_folder
    self.image_paths = [os.path.join(image_folder, f) for f in os.listdir(image_folder)]
    self.transform = transform

  def __len__(self):
    return len(self.image_paths)

  def __getitem__(self, index):
    image_path = self.image_paths[index]
    image = Image.open(image_path).covert
    if self.transform:
      image = self.transform(image)
    return image
# Defining transformations
transform = transforms.Compose([
    transforms.Resize((256, 256)),
  transforms.ToTensor(),
])

# Creating dataset and dataloader
dataset = MemeDataset("memes", transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

# Building the Meme Generator

Combining the trained models to generate memes:

- Use NLP model to generate captions
- Use the image model to select or generate meme templates.
- Overlay the captions on the image using Pillow library

```python
from PIL import Image, ImageDraw, ImageFont

# Overlay text on an image
def create_meme(image_path, caption):
  image = Image.open(image_path)
  draw = ImageDraw.Draw(image)
  font = ImageFont.load_default()
  draw.text((10, 10), caption, font=font, fill="black")
  return image

# Generating a meme
meme_image = create_meme("memes/meme_0.jpg", "First Caption, SOOO EGGGCITED")
meme_image.save("generated_meme.jpg")
```
